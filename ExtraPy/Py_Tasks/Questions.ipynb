1. Write a Python program to take a string input and count the number of vowels and consonants.


```python
def count_vowels_consonants():
   
    try:
        text = input("Enter a string: ").lower().strip()
        
        if not text:
            print("Error: Empty string entered!")
            return
    
        vowels = "aeiou"
        
        vowel_count = 0
        consonant_count = 0
              
        for char in text:
            if char.isalpha():  # Check if character is a letter
                if char in vowels:
                    vowel_count += 1
                else:
                    consonant_count += 1
        
    
        print(f"\nResults for string: '{text}'")
        print(f"Vowels: {vowel_count}")
        print(f"Consonants: {consonant_count}")
        print(f"Total letters: {vowel_count + consonant_count}")
        
    except Exception as e:
        print(f"An error occurred: {e}")
if __name__ == "__main__":
    print("=== Vowel and Consonant Counter ===")
    count_vowels_consonants()
```

2. Pattern Challenge: Print the star down pattern for n=5:


```python
def print_pattern(n=5):
  
    try:
        if not isinstance(n, int):
            print("Error: Input must be an integer!")
            return
        
        if n <= 0:
            print("Error: Input must be a positive integer!")
            return
        
        print(f"\nPattern for n = {n}:")
    
        for i in range(1, n + 1):
            print('*' * i)
            
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    print("Pattern Printer")
    print_pattern()
    
    print("\n" + "="*30 + "\n")
    
    try:
        user_n = int(input("Enter number of rows for pattern: "))
        print_pattern(user_n)
    except ValueError:
        print("Error: Please enter a valid integer!")
```

3. Write a Python function that accepts a string and returns the string with vowels removed.

```python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    
    # Remove vowels using list comprehension
    result = ''.join([char for char in input_string if char not in vowels])
    
    return result

def test_remove_vowels():
    print("Vowel Remover")
    
    test_cases = [
        "Hello World!",
        "Python Programming",
        "AEIOUaeiou",
        "12345!@#$%",
        "The quick brown fox",
        "",
        "BCDFG"
    ]
    
    for test_str in test_cases:
        print(f"\nOriginal: '{test_str}'")
        result = remove_vowels(test_str)
        print(f"Without vowels: '{result}'")
    
    # Interactive
    print("\n" + "="*40)
    user_input = input("\nEnter your own string to remove vowels: ")
    result = remove_vowels(user_input)
    print(f"\nOriginal: '{user_input}'")
    print(f"Without vowels: '{result}'")

if __name__ == "__main__":
    test_remove_vowels()
```

4. Write a Python class Rectangle with methods to calculate area and perimeter. Create objects and test them.

```python
class Rectangle:

    def __init__(self, length, width):
    
        try:
            self.length = float(length)
            self.width = float(width)
            
            # Validate dimensions
            if self.length <= 0 or self.width <= 0:
                raise ValueError("Dimensions must be positive numbers")
                
        except ValueError as e:
            print(f"Error creating rectangle: {e}")
            self.length = 1.0
            self.width = 1.0
    
    def calculate_area(self):
        return self.length * self.width
    
    def calculate_perimeter(self):
        return 2 * (self.length + self.width)
    
    def display_info(self):
    
        print(f"\nRectangle:")
        print(f"  Length: {self.length}")
        print(f"  Width: {self.width}")
        print(f"  Area: {self.calculate_area():.2f}")
        print(f"  Perimeter: {self.calculate_perimeter():.2f}")
    
    def is_square(self):

        return self.length == self.width

def test_rectangle_class():
    print("=== Rectangle Class Tester ===")
    
    test_rectangles = [
        Rectangle(5, 3),
        Rectangle(4, 4),
        Rectangle(7.5, 2.5),
        Rectangle(10, 6),
        Rectangle(0, 5),  # trigger error handling
        Rectangle(-2, 3)  # trigger error handling
    ]
    
    for i, rect in enumerate(test_rectangles, 1):
        print(f"\n{'='*40}")
        print(f"Rectangle {i}:")
        rect.display_info()
        if rect.is_square():
            print("  This rectangle is a square!")
    
    # Interactive test
    print("\n" + "="*40)
    print("\nCreate your own rectangle:")
    
    try:
        user_length = float(input("Enter length: "))
        user_width = float(input("Enter width: "))
        
        user_rect = Rectangle(user_length, user_width)
        user_rect.display_info()
        
        if user_rect.is_square():
            print("  This rectangle is a square!")
            
    except ValueError:
        print("Error: Please enter valid numbers!")

if __name__ == "__main__":
    test_rectangle_class()
```

5. Write a Python program that appends user input to a file until the user types "exit".

```python
def append_to_file():
    
    filename = "user_inputs.txt"
    
    print("File Appending")
    print("Type 'exit' to quit the program.")
    print(f"All inputs will be saved to '{filename}'")
    print("-" * 40)
    
    try:
        # Open file in append mode (creates file if it doesn't exist)
        with open(filename, 'a', encoding='utf-8') as file:
            file.seek(0, 2)  
            if file.tell() == 0:  
                file.write("User Input Log\n")
                file.write("=" * 40 + "\n\n")
            
            input_count = 0
            
            while True:
                user_input = input(f"\nEnter text [{input_count + 1}]: ")
                
        
                if user_input.lower() == 'exit':
                    print("\nExiting program. Goodbye!")
                    break
                
                # Append to file with timestamp
                from datetime import datetime
                timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                
                file.write(f"[{timestamp}] Entry {input_count + 1}: {user_input}\n")
                file.flush()  # Ensure immediate write to file
                
                print(f"âœ“ Saved to '{filename}'")
                input_count += 1
        
        print(f"\n{'='*40}")
        print(f"Summary:")
        print(f"Total entries saved: {input_count}")
        print(f"File: {filename}")
        
        # Show file contents if requested
        if input_count > 0:
            show_contents = input("\nWould you like to see the file contents? (yes/no): ")
            if show_contents.lower() == 'yes':
                print(f"\nContents of '{filename}':")
                print("-" * 40)
                with open(filename, 'r', encoding='utf-8') as file:
                    print(file.read())
    
    except PermissionError:
        print(f"Error: Cannot write to '{filename}'. Check file permissions.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        print("\nProgram execution completed.")

def view_file_contents():
    filename = "user_inputs.txt"
    
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            print(f"\nContents of '{filename}':")
            print("=" * 40)
            print(file.read())
    except FileNotFoundError:
        print(f"File '{filename}' not found. Run the program first to create it.")
    except Exception as e:
        print(f"Error reading file: {e}")

if __name__ == "__main__":
    append_to_file()
    
    # Option to view file contents after execution
    print("\n" + "="*40)
    view_option = input("\nWould you like to view the file contents now? (yes/no): ")
    if view_option.lower() == 'yes':
        view_file_contents()
```
