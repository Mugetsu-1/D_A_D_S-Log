1. Write a Python program to take a string input and count the number of vowels and consonants.


```python
def count_vowels_consonants():
   
    try:
        text = input("Enter a string: ").lower().strip()
        
        if not text:
            print("Error: Empty string entered!")
            return
    
        vowels = "aeiou"
        
        vowel_count = 0
        consonant_count = 0
              
        for char in text:
            if char.isalpha():  # Check if character is a letter
                if char in vowels:
                    vowel_count += 1
                else:
                    consonant_count += 1
        
    
        print(f"\nResults for string: '{text}'")
        print(f"Vowels: {vowel_count}")
        print(f"Consonants: {consonant_count}")
        print(f"Total letters: {vowel_count + consonant_count}")
        
    except Exception as e:
        print(f"An error occurred: {e}")
if __name__ == "__main__":
    print("=== Vowel and Consonant Counter ===")
    count_vowels_consonants()
```

2. Pattern Challenge: Print the star down pattern for n=5:


```python
def print_pattern(n=5):
  
    try:
        if not isinstance(n, int):
            print("Error: Input must be an integer!")
            return
        
        if n <= 0:
            print("Error: Input must be a positive integer!")
            return
        
        print(f"\nPattern for n = {n}:")
    
        for i in range(1, n + 1):
            print('*' * i)
            
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    print("Pattern Printer")
    print_pattern()
    
    print("\n" + "="*30 + "\n")
    
    try:
        user_n = int(input("Enter number of rows for pattern: "))
        print_pattern(user_n)
    except ValueError:
        print("Error: Please enter a valid integer!")
```
