1. Create a NumPy array of numbers from 10 to 50.

```python
import numpy as np

def create_array_10_to_50():
    
    print("Array from 10 to 50\n")
    
    # Using arange
    arr1 = np.arange(10, 51)
    print(f"Using np.arange(10, 51):\n{arr1}")
    print(f"Length: {len(arr1)}")
    print(f"Type: {type(arr1)}")
    
    # Using range with conversion
    arr2 = np.array(range(10, 51))
    print(f"\nUsing np.array(range(10, 51)):\n{arr2}")
    
    return arr1

array_10_50 = create_array_10_to_50()
```

2. Generate a 3×3 identity matrix.

```python
def create_identity_matrix():
    print("\n" + "="*40 + "\n")
    print("3×3 Identity Matrix\n")
    
    # Create identity matrix
    identity_matrix = np.eye(3)
    print(f"3×3 Identity Matrix:\n{identity_matrix}")
    print(f"\nShape: {identity_matrix.shape}")
    print(f"Data type: {identity_matrix.dtype}")
    
    print(f"\nVerification:")
    print(f"Sum of diagonal elements: {np.trace(identity_matrix)}")
    print(f"Is identity matrix? {np.allclose(identity_matrix, np.eye(3))}")
    
    # Show alternative method
    identity_alt = np.identity(3)
    print(f"\nAlternative using np.identity(3):\n{identity_alt}")
    
    return identity_matrix
identity_3x3 = create_identity_matrix()
```

3. Create a NumPy array of 10 random integers between 1 and 100.

```python
def create_random_array():
    print("\n" + "="*40 + "\n")
    print("Random Integers (1-100)\n")
    
    # Set seed for reproducibility
    np.random.seed(42)
    random_array = np.random.randint(1, 101, 10)
    
    print(f"10 Random integers between 1 and 100:")
    print(random_array)
    print(f"\nStatistics:")
    print(f"Sum: {random_array.sum()}")
    print(f"Mean: {random_array.mean():.2f}")
    print(f"Median: {np.median(random_array)}")
    print(f"Standard Deviation: {random_array.std():.2f}")
    
    print(f"\nWithout fixed seed (different each run):")
    truly_random = np.random.randint(1, 101, 10)
    print(truly_random)
    
    return random_array
random_ints = create_random_array()
```
