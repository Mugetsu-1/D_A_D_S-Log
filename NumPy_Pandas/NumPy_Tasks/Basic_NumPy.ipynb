2. Generate a 3×3 identity matrix.

```python
def create_identity_matrix():
    print("\n" + "="*40 + "\n")
    print("3×3 Identity Matrix\n")
    
    # Create identity matrix
    identity_matrix = np.eye(3)
    print(f"3×3 Identity Matrix:\n{identity_matrix}")
    print(f"\nShape: {identity_matrix.shape}")
    print(f"Data type: {identity_matrix.dtype}")
    
    print(f"\nVerification:")
    print(f"Sum of diagonal elements: {np.trace(identity_matrix)}")
    print(f"Is identity matrix? {np.allclose(identity_matrix, np.eye(3))}")
    
    # Show alternative method
    identity_alt = np.identity(3)
    print(f"\nAlternative using np.identity(3):\n{identity_alt}")
    
    return identity_matrix
identity_3x3 = create_identity_matrix()
```

4. Find the maximum and minimum of an array.

```python
def find_max_min():
    
    print("\n" + "="*40 + "\n")
    print("Maximum and Minimum\n")
    arr = np.array([23, 45, 12, 89, 34, 67, 98, 5, 41, 72])
    print(f"Sample array:\n{arr}")
    
    # Find maximum and minimum
    max_val = arr.max()
    min_val = arr.min()
    
    # Alternative methods
    max_val_alt = np.max(arr)
    min_val_alt = np.min(arr)
    
    print(f"\nUsing array methods:")
    print(f"Maximum value: {max_val}")
    print(f"Minimum value: {min_val}")
    
    print(f"\nUsing numpy functions:")
    print(f"Maximum value: {max_val_alt}")
    print(f"Minimum value: {min_val_alt}")
    
    # Find indices of max and min
    max_index = arr.argmax()
    min_index = arr.argmin()
    
    print(f"\nIndices:")
    print(f"Index of maximum value ({max_val}): {max_index}")
    print(f"Index of minimum value ({min_val}): {min_index}")
    
    # For 2D array example
    print(f"\n" + "-"*30)
    print("For 2D array:")
    arr_2d = np.random.randint(1, 50, (3, 4))
    print(f"2D array:\n{arr_2d}")
    print(f"Global max: {arr_2d.max()}")
    print(f"Global min: {arr_2d.min()}")
    print(f"Max in each column: {arr_2d.max(axis=0)}")
    print(f"Min in each row: {arr_2d.min(axis=1)}")
    
    return arr, max_val, min_val

sample_array, max_value, min_value = find_max_min()
```

6. Create an array of all even numbers between 50 and 100.

```python
def even_numbers_array():
    print("\n" + "="*40 + "\n")
    print("Even Numbers (50-100)\n")
    
    # Using arange with step
    even_arr1 = np.arange(50, 101, 2)
    print(f"Method 1 (np.arange with step=2):")
    print(even_arr1)
    print(f"Number of elements: {len(even_arr1)}")
    
    # Using filtering
    all_numbers = np.arange(50, 101)
    even_arr2 = all_numbers[all_numbers % 2 == 0]
    print(f"\nMethod 2 (filtering with condition):")
    print(even_arr2)
    
    # List comprehension
    even_arr3 = np.array([x for x in range(50, 101) if x % 2 == 0])
    print(f"\nMethod 3 (list comprehension):")
    print(even_arr3)
    
    # Verify all methods give same result
    print(f"\nVerification:")
    print(f"Method 1 == Method 2: {np.array_equal(even_arr1, even_arr2)}")
    print(f"Method 2 == Method 3: {np.array_equal(even_arr2, even_arr3)}")

    print(f"\nStatistics of even numbers:")
    print(f"Sum: {even_arr1.sum()}")
    print(f"Mean: {even_arr1.mean():.2f}")
    print(f"Standard Deviation: {even_arr1.std():.2f}")
    
    return even_arr1
even_array = even_numbers_array()
```

9. Find the index of the maximum element in [10, 20, 50, 40, 30].

```python
def find_max_index():
    print("\n" + "="*40 + "\n")
    print("Index of Maximum Element\n")
  
    arr = np.array([10, 20, 50, 40, 30])
    print(f"Array: {arr}")
    
    # Find maximum value and its index
    max_value = arr.max()
    max_index = arr.argmax()
    
    print(f"\nMaximum value: {max_value}")
    print(f"Index of maximum value: {max_index}")
  
    print(f"\nAlternative methods:")
    max_index_alt1 = np.argmax(arr)
    print(f"Using np.argmax(): {max_index_alt1}")
    
    # Using where for multiple maxima
    max_indices = np.where(arr == max_value)[0]
    print(f"Using np.where() (useful for multiple maxima): {max_indices}")
 
    print(f"\n" + "-"*30)
    print("Example with duplicate maximum values:")
    arr_duplicate = np.array([10, 50, 30, 50, 20])
    print(f"Array: {arr_duplicate}")
    
    max_val_dup = arr_duplicate.max()
    max_indices_dup = np.where(arr_duplicate == max_val_dup)[0]
    
    print(f"Maximum value: {max_val_dup}")
    print(f"Indices of maximum value: {max_indices_dup}")
    
    # For 2D array
    print(f"\n" + "-"*30)
    print("For 2D array:")
    arr_2d = np.array([[1, 5, 3], [7, 2, 9], [4, 8, 6]])
    print(f"2D array:\n{arr_2d}")
    
    max_index_2d = np.unravel_index(np.argmax(arr_2d), arr_2d.shape)
    print(f"Index of maximum in 2D array (row, col): {max_index_2d}")
    
    return arr, max_value, max_index

test_array, max_val, max_idx = find_max_index()
```

10. Create two arrays and perform element-wise addition, subtraction, multiplication, and division.

```python
def elementwise_operations():
    print("\n" + "="*40 + "\n")
    print("Element-wise Array Operations\n")
    
    # Create two arrays
    np.random.seed(42)
    arr1 = np.random.randint(1, 11, 5)
    arr2 = np.random.randint(1, 11, 5)  
    
    print(f"Array 1: {arr1}")
    print(f"Array 2: {arr2}")
    print(f"\nBoth arrays have same shape: {arr1.shape == arr2.shape}")
    
    # Element-wise operations
    print(f"\n" + "="*30)
    print("Element-wise Operations:")
    
    # Addition
    addition = arr1 + arr2
    print(f"\n1. Addition (arr1 + arr2):")
    print(f"   {arr1} + {arr2} = {addition}")
    
    # Subtraction
    subtraction = arr1 - arr2
    print(f"\n2. Subtraction (arr1 - arr2):")
    print(f"   {arr1} - {arr2} = {subtraction}")
    
    # Multiplication
    multiplication = arr1 * arr2
    print(f"\n3. Multiplication (arr1 * arr2):")
    print(f"   {arr1} * {arr2} = {multiplication}")
    
    # Division (with zero handling)
    print(f"\n4. Division (arr1 / arr2):")
    # Create arrays with no zeros for division
    arr3 = np.array([10, 20, 30, 40, 50], dtype=float)
    arr4 = np.array([2, 4, 5, 8, 10], dtype=float)
    
    print(f"   New arrays for division (to avoid division by zero):")
    print(f"   Array 3: {arr3}")
    print(f"   Array 4: {arr4}")
    
    division = arr3 / arr4
    print(f"   Division result: {division}")
    
    # Power operation
    power = arr1 ** 2
    print(f"\n5. Power (arr1 ** 2):")
    print(f"   {arr1} ** 2 = {power}")
    
    # Modulus
    modulus = arr1 % arr2
    print(f"\n6. Modulus (arr1 % arr2):")
    print(f"   {arr1} % {arr2} = {modulus}")
    
    # Using numpy functions
    print(f"\n" + "="*30)
    print("Using NumPy functions:")
    
    print(f"\nnp.add(arr1, arr2): {np.add(arr1, arr2)}")
    print(f"np.subtract(arr1, arr2): {np.subtract(arr1, arr2)}")
    print(f"np.multiply(arr1, arr2): {np.multiply(arr1, arr2)}")
    print(f"np.divide(arr3, arr4): {np.divide(arr3, arr4)}")
    
    # Broadcasting example
    print(f"\n" + "="*30)
    print("Broadcasting Example:")
    scalar = 5
    print(f"Array: {arr1}")
    print(f"Scalar: {scalar}")
    print(f"Array + scalar: {arr1 + scalar}")
    print(f"Array * scalar: {arr1 * scalar}")
    
    return arr1, arr2, arr3, arr4

a1, a2, a3, a4 = elementwise_operations()
```
