1. Create an array of numbers from 1 to 10 and print only even numbers

```python
import numpy as np

def extract_even_numbers():
    
    print("Extract Even Numbers (1-10)\n")
    arr = np.arange(1, 11)
    print(f"Original array (1-10): {arr}")
    
    # Using modulo operator
    even_numbers = arr[arr % 2 == 0]
    print(f"\nMethod 1 - Using modulo (arr % 2 == 0):")
    print(f"Even numbers: {even_numbers}")
    
    # Using fancy indexing
    even_indices = [1, 3, 5, 7, 9] 
    even_numbers_alt = arr[even_indices]
    print(f"\nMethod 2 - Using fancy indexing:")
    print(f"Even numbers: {even_numbers_alt}")
    
    # Using arrange with step
    even_numbers_step = np.arange(2, 11, 2)
    print(f"\nMethod 3 - Direct creation with np.arange(2, 11, 2):")
    print(f"Even numbers: {even_numbers_step}")
    
    # Verification
    print(f"\nVerification:")
    print(f"All methods give same result: {np.array_equal(even_numbers, even_numbers_alt) and np.array_equal(even_numbers_alt, even_numbers_step)}")
    
    return arr, even_numbers
original_arr, evens = extract_even_numbers()
```

2. Extract the last 3 elements from the array

```python
def extract_last_elements():
    print("\n" + "="*40 + "\n")
    print("Extract Last 3 Elements\n")
    arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
    print(f"Sample array: {arr}")
    print(f"Array length: {len(arr)}")
    
    # Using negative indexing
    last_three = arr[-3:]
    print(f"\nMethod 1 - Using negative indexing (arr[-3:]):")
    print(f"Last 3 elements: {last_three}")
    
    # Using slice with length
    last_three_alt = arr[len(arr)-3:]
    print(f"\nMethod 2 - Using slice with length (arr[len(arr)-3:]):")
    print(f"Last 3 elements: {last_three_alt}")
    
    # Using np.take
    last_three_take = np.take(arr, [-3, -2, -1])
    print(f"\nMethod 3 - Using np.take (np.take(arr, [-3, -2, -1])):")
    print(f"Last 3 elements: {last_three_take}")
    
    # Extract last n elements function
    def get_last_n(arr, n):
      
        if n > len(arr):
            return arr
        return arr[-n:]
  
    print(f"\nTesting with different values of n:")
    for n in [1, 3, 5, 7]:
        result = get_last_n(arr, n)
        print(f"Last {n} element(s): {result}")
    
    return arr, last_three

sample_arr, last_three = extract_last_elements()
```

3. Array Operations on Sample Array

```python
def array_operations():
   
    print("\n" + "="*40 + "\n")
    print("Array Operations\n")

    arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
    print(f"Original array: {arr}")
    print(f"Array shape: {arr.shape}")
    print(f"Array indices: {list(range(len(arr)))}")
    
    # Task 1: Slice and print every third element starting from index 1
    print("\n" + "="*40)
    print("Task 1: Every third element starting from index 1")
    
    # (every third element)
    third_elements = arr[1::3]
    print(f"Indices selected: {list(range(1, len(arr), 3))}")
    print(f"Values: {third_elements}")
    print(f"Explanation: arr[1::3] means start at index 1, go to end, step by 3")
    
    # Task 2: Print the last 4 elements using negative indexing
    print("\n" + "="*40)
    print("Task 2: Last 4 elements using negative indexing")
    
    last_four = arr[-4:]
    print(f"Negative indices: {list(range(-4, 0))}")
    print(f"Corresponding positive indices: {list(range(len(arr)-4, len(arr)))}")
    print(f"Last 4 elements: {last_four}")
    print(f"Explanation: arr[-4:] means from 4th from end to the end")
    
    # Task 3: Replace the values at even indices with -1
    print("\n" + "="*40)
    print("Task 3: Replace values at even indices with -1")
    
    # Duplicate to avid alter in Original array
    arr_copy = arr.copy()
    print(f"Original array (copy): {arr_copy}")

    even_indices = np.arange(0, len(arr_copy), 2)
    print(f"Even indices: {even_indices}")
    
    # Replace values at even indices
    arr_copy[even_indices] = -1
    print(f"After replacement: {arr_copy}")
    
    # Task 4: Use fancy indexing to extract values at positions [2, 4, 6, 8]
    print("\n" + "="*40)
    print("Task 4: Extract values at positions [2, 4, 6, 8]")
    
    fancy_indices = [2, 4, 6, 8]
    fancy_values = arr[fancy_indices]
    print(f"Indices: {fancy_indices}")
    print(f"Corresponding values: {fancy_values}")
    print(f"Explanation: arr[[2, 4, 6, 8]] extracts elements at these specific indices")
    
    # Summary
    print("\n" + "="*40)
    print("SUMMARY OF OPERATIONS")
    print("="*40)
    
    summary = [
        ["Operation", "Syntax", "Result"],
        ["Original array", "arr", str(arr)],
        ["Every 3rd from index 1", "arr[1::3]", str(third_elements)],
        ["Last 4 elements", "arr[-4:]", str(last_four)],
        ["Even indices replaced", "arr_copy[even_indices] = -1", str(arr_copy)],
        ["Fancy indexing [2,4,6,8]", "arr[[2,4,6,8]]", str(fancy_values)]
    ]
    
    for row in summary:
        print(f"{row[0]:<25} | {row[1]:<20} | {row[2]}")
    
    return arr, third_elements, last_four, arr_copy, fancy_values

orig_array, third_els, last_4, modified_arr, fancy_vals = array_operations()
```

4. Reshape and Flatten Operations

```python
def reshape_flatten_operations():

    print("\n" + "="*40 + "\n")
    print("Reshape and Flatten Operations\n")
    
    # Create a 1D array with 48 elements (for 6x8 reshape)
    arr_1d = np.arange(1, 49)
    print(f"1. Original 1D array (first 10 elements):")
    print(f"   {arr_1d[:10]} ... {arr_1d[-10:]}")
    print(f"   Shape: {arr_1d.shape}, Total elements: {len(arr_1d)}")
    
    # Reshape to 6x8
    print("\n2. Reshape to 6x8 matrix:")
    arr_6x8 = arr_1d.reshape(6, 8)
    print(f"   Shape: {arr_6x8.shape}")
    print(f"   Matrix:\n{arr_6x8}")
    
    # Check reshaping rules
    print(f"\n   Verification:")
    print(f"   Can reshape to 6x8? {arr_1d.size == 6*8}")
    print(f"   Total elements: {arr_6x8.size}")
    
    # Flatten the 6x8 matrix back to 1D
    print("\n3. Flatten 6x8 matrix back to 1D:")
    arr_flattened = arr_6x8.flatten()
    print(f"   Shape: {arr_flattened.shape}")
    print(f"   First 10 elements: {arr_flattened[:10]}")
    print(f"   Equal to original? {np.array_equal(arr_1d, arr_flattened)}")
    
    # Alternative flattening methods
    print("\n4. Alternative flattening methods:")
    
    # ravel()
    arr_ravel = arr_6x8.ravel()
    print(f"   Using ravel(): {arr_ravel[:10]}...")
    print(f"   ravel() == flatten()? {np.array_equal(arr_ravel, arr_flattened)}")
    
    # reshape(-1)
    arr_reshape = arr_6x8.reshape(-1)
    print(f"   Using reshape(-1): {arr_reshape[:10]}...")
    print(f"   reshape(-1) == flatten()? {np.array_equal(arr_reshape, arr_flattened)}")
    
    # Differences between methods
    print("\n5. Differences between flatten methods:")
    print("   - flatten(): Returns a copy (new array)")
    print("   - ravel(): Returns a view (if possible) or copy")
    print("   - reshape(-1): Returns a view (if possible)")
    
    # Test if it's a view or copy
    test_array = arr_6x8.copy()
    
    # Using flatten (creates copy)
    flat_copy = test_array.flatten()
    flat_copy[0] = 999
    print(f"\n   Test - flatten creates copy:")
    print(f"   Modified flatten[0] = 999, original[0,0] = {test_array[0,0]}")
    
    # Using ravel (creates view when possible)
    ravel_view = test_array.ravel()
    ravel_view[0] = 888
    print(f"\n   Test - ravel creates view (modifies original):")
    print(f"   Modified ravel[0] = 888, original[0,0] = {test_array[0,0]}")
    
    return arr_1d, arr_6x8, arr_flattened

array_1d, array_6x8, flattened_array = reshape_flatten_operations()
```

5. Convert 3D Array Directly to 1D Array using Reshape

```python
def convert_3d_to_1d():
    print("\n" + "="*40 + "\n")
    print("Convert 3D Array to 1D using Reshape\n")
    
    # Create a 3D array (2x3x4)
    print("1. Create a 3D array (2x3x4):")
    arr_3d = np.arange(1, 25).reshape(2, 3, 4)
    print(f"   Shape: {arr_3d.shape}")
    print(f"   Total elements: {arr_3d.size}")
    
    print("\n   The 3D array consists of 2 matrices, each 3x4:")
    for i in range(arr_3d.shape[0]):
        print(f"\n   Matrix {i+1}:")
        print(f"   {arr_3d[i]}")
    
    # Convert directly to 1D using reshape(-1)
    print("\n2. Convert directly to 1D using reshape(-1):")
    arr_1d = arr_3d.reshape(-1)
    print(f"   Shape: {arr_1d.shape}")
    print(f"   1D array: {arr_1d}")
    
    # using flatten()
    print("\n3. Alternative using flatten():")
    arr_1d_flat = arr_3d.flatten()
    print(f"   Shape: {arr_1d_flat.shape}")
    print(f"   1D array: {arr_1d_flat}")
    
    print(f"\n4. Verification:")
    print(f"   reshape(-1) == flatten()? {np.array_equal(arr_1d, arr_1d_flat)}")
    
    # Show reshaping
    print("\n5. Understanding the reshaping process:")
    print("   Original 3D shape: (2, 3, 4)")
    print("   This means: 2 blocks, each with 3 rows and 4 columns")
    print("   Total elements: 2 × 3 × 4 = 24")
    print("   Reshape to 1D preserves the order:")
    print("   Order: arr_3d[0,0,0], arr_3d[0,0,1], ..., arr_3d[1,2,3]")
    
    # verification of order
    print("\n6. Manual verification of element order:")
    print("   First few elements from 3D array:")
    print(f"   arr_3d[0,0,0] = {arr_3d[0,0,0]}")
    print(f"   arr_3d[0,0,1] = {arr_3d[0,0,1]}")
    print(f"   arr_3d[0,0,2] = {arr_3d[0,0,2]}")
    print(f"   arr_3d[0,0,3] = {arr_3d[0,0,3]}")
    print(f"   ...")
    print(f"   arr_3d[1,2,3] = {arr_3d[1,2,3]}")
    
    # Convert to different 1D shapes
    print("\n7. Other 1D conversions:")
    
    # Row-major order 
    arr_1d_c = arr_3d.reshape(-1, order='C')
    print(f"   C-style (row-major): {arr_1d_c}")
    
    # (Fortran-style)
    arr_1d_f = arr_3d.reshape(-1, order='F')
    print(f"\n   F-style (column-major): {arr_1d_f}")
    
    print(f"\n   First 6 elements comparison:")
    print(f"   C-style: {arr_1d_c[:6]}")
    print(f"   F-style: {arr_1d_f[:6]}")
    
    return arr_3d, arr_1d, arr_1d_c, arr_1d_f
array_3d, array_1d, array_1d_c, array_1d_f = convert_3d_to_1d()
```
