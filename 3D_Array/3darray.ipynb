```python
import numpy as np
np.random.seed(42)  # For reproducible results
```

1. Create a 3x3 matrix with values from 1 to 9

```python
# Using reshape
matrix_3x3 = np.arange(1, 10).reshape(3, 3)
print("3x3 Matrix (1-9):")
print(matrix_3x3)
print(f"Shape: {matrix_3x3.shape}\n")
```

2. Create an array of 10 random floats between 0 and 1, rounded to 2 decimal places

```python
random_floats = np.round(np.random.rand(10), 2)
print("10 Random floats (0-1):")
print(random_floats)
print(f"Shape: {random_floats.shape}\n")
```

3. Find all even numbers in the array using Boolean indexing

```python
arr = np.array([1, 3, 6, 9, 12, 15])
even_numbers = arr[arr % 2 == 0]
print(f"Original array: {arr}")
print(f"Even numbers: {even_numbers}\n")
```

4. Create a 4x4 matrix filled with 7

```python
matrix_4x4 = np.full((4, 4), 7)
print("4x4 Matrix filled with 7:")
print(matrix_4x4)
print(f"Shape: {matrix_4x4.shape}\n")
```

5. Create a 3D array of shape (2, 3, 4) with random integers between 1 and 100

```python
array_3d = np.random.randint(1, 101, size=(2, 3, 4))
print("3D Array (2, 3, 4) with random integers 1-100:")
print(array_3d)
print(f"Shape: {array_3d.shape}\n")
```

6. Extract all elements greater than 50 and replace them with -1

```python
# 3D array from question 5
array_6 = array_3d.copy()  # copy to save original
mask_greater_than_50 = array_6 > 50
print("Original array:")
print(array_6)

print("\nElements greater than 50:")
print(array_6[mask_greater_than_50])

# Replace elements > 50 with -1
array_6[mask_greater_than_50] = -1
print("\nArray after replacing >50 with -1:")
print(array_6)
print()
```

7. 3D Array Operations

```python
# Create a 3D array of shape (3, 4, 5) with random integers between 1 and 100
array_3d_large = np.random.randint(1, 101, size=(3, 4, 5))
print("Original 3D Array (3, 4, 5):")
print(array_3d_large)
print(f"Shape: {array_3d_large.shape}")

mean_value = np.mean(array_3d_large)
print(f"\nMean of entire array: {mean_value:.2f}")

# Replace all elements greater than the mean with 0
array_replaced = array_3d_large.copy()
array_replaced[array_replaced > mean_value] = 0
print("\nArray after replacing elements > mean with 0:")
print(array_replaced)

# Calculate the sum along axis=1
sum_axis_1 = np.sum(array_3d_large, axis=1)
print("\nSum along axis=1 (sum across 5 columns in each depth slice):")
print(sum_axis_1)
print(f"Shape of sum_axis_1: {sum_axis_1.shape}")

# Extract first and last columns from each row of each depth slice
first_last_cols = array_3d_large[:, :, [0, -1]]
print("\nFirst and last columns from each row of each depth slice:")
print(first_last_cols)
print(f"Shape: {first_last_cols.shape}")
```
