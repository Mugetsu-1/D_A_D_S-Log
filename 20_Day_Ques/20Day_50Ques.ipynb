# 41. Normalize NumPy array (mean = 0, std = 1)
def normalize_array(arr):
    """Normalize array to have mean 0 and standard deviation 1"""
    mean = np.mean(arr)
    std = np.std(arr)
    if std == 0:  # Avoid division by zero
        return arr - mean
    return (arr - mean) / std

# Test
original_array = np.array([1, 2, 3, 4, 5])
normalized = normalize_array(original_array)
print("Original array:", original_array)
print(f"Mean of original: {np.mean(original_array):.2f}, Std: {np.std(original_array):.2f}")
print("Normalized array:", normalized)
print(f"Mean of normalized: {np.mean(normalized):.2f}, Std: {np.std(normalized):.2f}")

# 45. Filter elements greater than average
def filter_above_average(arr):
    """Filter elements greater than average"""
    avg = np.mean(arr)
    return arr[arr > avg]

# 49. Flatten 2D NumPy array without using .flatten()
def flatten_array(arr):
    """Flatten 2D array without using .flatten()"""
    flattened = []
    for row in arr:
        for element in row:
            flattened.append(element)
    return np.array(flattened)

# Test
matrix = np.array([[1, 2, 3], [4, 5, 6]])
flattened = flatten_array(matrix)
print("Original matrix:")
print(matrix)
print("\nFlattened array:", flattened)

# Alternative using reshape
flattened_reshape = matrix.reshape(-1)
print("\nUsing reshape method:", flattened_reshape)
