47. Replace maximum element with 0 in random matrix

```python
matrix = np.random.rand(4, 4)
print("Original matrix:")
print(matrix)

max_value = np.max(matrix)
matrix[matrix == max_value] = 0

print("\nAfter replacing maximum with 0:")
print(matrix)
```

48. Find unique elements and their counts in NumPy array

```python
array = np.array([1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5])
unique_elements, counts = np.unique(array, return_counts=True)

print("Array:", array)
print("\nUnique elements and counts:")
for element, count in zip(unique_elements, counts):
    print(f"  {element}: {count}")
```

49. Flatten 2D NumPy array without using .flatten()

```python
def flatten_array(arr):

    flattened = []
    for row in arr:
        for element in row:
            flattened.append(element)
    return np.array(flattened)

# Test
matrix = np.array([[1, 2, 3], [4, 5, 6]])
flattened = flatten_array(matrix)
print("Original matrix:")
print(matrix)
print("\nFlattened array:", flattened)

# Using reshape
flattened_reshape = matrix.reshape(-1)
print("\nUsing reshape method:", flattened_reshape)
```

50. Extract diagonal elements from 5x5 array

```python
array = np.arange(25).reshape(5, 5)
diagonal = np.diag(array)

print("5x5 array (0-24):")
print(array)
print("\nDiagonal elements:", diagonal)
```
