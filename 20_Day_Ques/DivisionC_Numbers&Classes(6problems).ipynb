10. Check if a number is an Armstrong number

```python
def is_armstrong_number(num):
    num_str = str(num)
    num_digits = len(num_str)
    
    sum_of_powers = sum(int(digit) ** num_digits for digit in num_str)
    
    return sum_of_powers == num

# Test
print("Is 153 Armstrong?", is_armstrong_number(153))
print("Is 123 Armstrong?", is_armstrong_number(123))
```

12. Generate Fibonacci sequence using recursion

```python
def fibonacci_recursive(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib_seq = fibonacci_recursive(n - 1)
        fib_seq.append(fib_seq[-1] + fib_seq[-2])
        return fib_seq

# Test
print("Fibonacci (10 terms):", fibonacci_recursive(10))
```

20. Get top 3 highest values from dictionary

```python
def top_three_values(dictionary):
    sorted_items = sorted(dictionary.items(), key=lambda x: x[1], reverse=True)
    return dict(sorted_items[:3])

# Test
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'David': 95, 'Eve': 88}
print("Top 3 scores:", top_three_values(scores))
```

23. Check if a number is prime

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    
    return True

# Test
print("Is 17 prime?", is_prime(17))
print("Is 20 prime?", is_prime(20))
```

29. Calculate sum of digits of a number

```python
def sum_of_digits(number):
    total = 0
    num = abs(number)  # for neg
    
    while num > 0:
        total += num % 10
        num //= 10
    
    return total

# Test
print("Sum of digits of 12345:", sum_of_digits(12345))
```

30. BankAccount class

```python
class BankAccount:
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder
        self.balance = initial_balance
        self.transactions = []
    
    def deposit(self, amount):
        if amount <= 0:
            print("Deposit amount must be positive")
            return
        
        self.balance += amount
        self.transactions.append(f"Deposit: +${amount}")
        print(f"Deposited ${amount}. New balance: ${self.balance}")
    
    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive")
            return
        
        if amount > self.balance:
            print("Insufficient funds")
            return
        
        self.balance -= amount
        self.transactions.append(f"Withdrawal: -${amount}")
        print(f"Withdrew ${amount}. New balance: ${self.balance}")
    
    def display_balance(self):
        print(f"Account holder: {self.account_holder}")
        print(f"Current balance: ${self.balance}")
    
    def display_transactions(self):
        print(f"Transaction history for {self.account_holder}:")
        for transaction in self.transactions:
            print(f"  {transaction}")

# Test
account = BankAccount("John Doe", 1000)
account.deposit(500)
account.withdraw(200)
account.display_balance()
account.display_transactions()
```
